```python
delay = flights[flights['DepDelay'] > 15 ]
delay[delay.isnull().any(axis=1)]
#Of the rows that has Null values in column DepDel15, none has a Departure Delay of more than 15, so we can continue to fill the null values with 0 as they were not late more than 15 minutes
```

```python
flights.DepDel15 = flights.DepDel15.fillna(0)
flights['DepDel15'].isnull().sum()
```

```python
print('Data is cleaned')
```

```python
DepDelay = flights['DepDelay'].describe()
ArrDelay = flights['ArrDelay'].describe()
print(DepDelay,ArrDelay)
#summary statistics for the 2 Columns
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
var_data = flights['DepDelay']

```

```python
def show_distribution(var_data):
    from matplotlib import pyplot as plt
    
    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean()
    med_val = var_data.median()
    mod_val = var_data.mode()[0]
    
    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(min_val,
                                                                                            mean_val,
                                                                                            med_val,
                                                                                            mod_val,
                                                                                            max_val
                                                                                                ))
    fig, ax = plt.subplots(2,1, figsize=(10,4))
    
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frequency')
    
    ax[0].axvline(x=min_val, color='grey', linestyle='dashed',linewidth=2)
    ax[0].axvline(x=mean_val, color='cyan', linestyle='dashed',linewidth=2)
    ax[0].axvline(x=med_val, color='red', linestyle='dashed',linewidth=2)
    ax[0].axvline(x=mod_val, color='yellow', linestyle='dashed',linewidth=2)
    ax[0].axvline(x=max_val, color='grey', linestyle='dashed',linewidth=2)
    
    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel('Value')
    
    fig.suptitle('Data Distribution')
    
    fig.show()
    
col = flights['DepDelay']

show_distribution(col)
```
