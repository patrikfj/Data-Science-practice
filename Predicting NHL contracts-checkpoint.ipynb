```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Contracts = pd.read_excel('NHL_player_contracts_2021.xlsx')
Player_statistics = pd.read_excel("NHL_playerstats_last3years.xlsx")
```

```python
Contracts.info()
#We have to change VALUE and CAP HIT from object to in64 so it can be used in ML
```

```python
Contracts.isnull().sum()
```

```python
Contracts['VALUE'] = Contracts['VALUE'].str.replace(',','').str.replace('$','')
Contracts['CAP HIT'] = Contracts['CAP HIT'].str.replace(',','').str.replace('$','')
Contracts = Contracts.astype({'VALUE':'int64'}).astype({'CAP HIT':'int64'})
Contracts.info()
#We changed VALUE and CAP HIT but we also needed to take out $ sign and comma in order to make it standardized
```

```python
var_data = (Contracts['VALUE'] )
fig = plt.figure(figsize=(10,4))
plt.hist(var_data)
fig.show
```

```python
Not_entrylvl = pd.Series(Contracts['CAP HIT'] >= 1000000)
Not_entrylvl.sum()
#of the 300 contracts, only 101 of them are not Entry level
#Since an entry level is usually give to players who have not played in the NHL, such as rookies and foreign player it is impossible to predict salary for someone who has not played in the NHL yet
Contracts = Contracts[Contracts['CAP HIT'] >= 1000000]
Contracts
```

```python
Player_statistics.info()
```

```python
Player_statistics.isnull().sum()
```

```python
Contracts = Contracts.rename(columns = {'PLAYER':'Player', 'S/C':'SHOOTS'})
Player_statistics = Player_statistics.rename(columns = {'S/C':'SHOOTS'})
```

```python
Stats_contract = pd.merge(Contracts, Player_statistics, how = "left", on = ['Player'])
#Merge the two dataframes on primary key 'Player' using left join
```

```python
Stats_contract.info()
```

```python
Stats_contract = Stats_contract.drop(columns=['TEAM','DATE','LENGTH','VALUE', 'EXTENSION']).dropna(thresh=16)
#Drop unessesary columns and drop players with 16 NaN values because if a player has NaN that means they do not have any stats from- 
# NHL (foreign players or college/juniors and is most likley just a entry level contract that is league manatory)
```

```python
Stats_contract.Pos.unique()
```

```python
def Position(Position):
    Pos = Position
    if Pos == 'D':
        return 0
    elif Pos == 'C':
        return 1
    else:
        return -1
    
def Structure(Structure):
    STRUCTURE = Structure
    if STRUCTURE == '1-way':
        return 1
    else:
        return 0
    
def Shoots(Shoots):
    SHOOTS = Shoots
    if SHOOTS == 'R':
        return 0
    else:
        return 1
```

```python
Stats_contract['Position'] = Stats_contract['Pos'].apply(Position)
Stats_contract['Structure'] = Stats_contract['STRUCTURE'].apply(Structure)
Stats_contract['Shoots'] = Stats_contract['SHOOTS'].apply(Shoots)
```

```python
Salary_pred = Stats_contract.drop(columns=['Player','TYPE','STRUCTURE','SHOOTS','Pos','POS','TOI/GP','S%','FOW%','SHG','SHP'])
Salary_pred
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, validation_curve, cross_val_score
from sklearn import preprocessing as pp
from sklearn.linear_model import LogisticRegression
```

```python
X = Salary_pred.loc[:, Salary_pred.columns != 'CAP HIT']
y = Salary_pred['CAP HIT']


max_abs_scaler = pp.MaxAbsScaler()
norm_X = max_abs_scaler.fit_transform(X)
norm_y = max_abs_scaler.fit_transform(y.values.reshape(-1,1))

norm_X_train, norm_X_test, norm_y_train, norm_y_test = train_test_split(norm_X, norm_y, test_size=0.25, random_state=42)

regr = linear_model.LinearRegression()
regr.fit(norm_X_train, norm_y_train)
Salary_y_pred = regr.predict(norm_X_test)
```

```python
print(round((regr.score(norm_X_test, norm_y_test)),2))
print(round(mean_squared_error(norm_y_test,Salary_y_pred),2))
scores = cross_val_score(regr, norm_X_test, norm_y_test, cv=4)
print(scores)
```

```python
from sklearn import tree
clf = tree.DecisionTreeRegressor()
clf = clf.fit(norm_X_train, norm_y_train)
Sal_pred = clf.predict(norm_X)
print(round((clf.score(norm_X_test,norm_y_test)),2))
scores = cross_val_score(clf, norm_X_test, norm_y_test, cv=4)
print(scores)
```
